### Задание 3.1
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.

Сделайте анализ в формате storytelling: дополните каждый график письменными выводами и наблюдениями.

```python
data_house = df[['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot',
           'floors', 'waterfront', 'view', 'condition', 'grade',
           'yr_built', 'yr_renovated']]
```

```python
corr_matrix = data_house.corr()
```

```python
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidths = 0.5, cmap='coolwarm');
```

Из тепловой карты видно, что сильная прямая существует между общей жилой площадью, количеством спален и ванных комнат.

<br>Существенное влияние на стоимость дома оказывают следующие факторы: количество жилой площади и дизайн дома. Также на цену дома оказывают влияние количество ванных комнат и общий вид дома. 


```python
plt.figure(figsize=(6, 4))

plt.scatter(df['grade'], df['price'])

plt.title('Зависимость стоимости дома от его дизайна')
plt.xlabel('Дизайн')
plt.ylabel('Стоимость дома');
```

Внешний вид дома (его дизайн) оказывыет влияние на его стоимость, т.к. чем выше цена дома, тем больше требований к его дизайну.

```python
plt.bar(df['condition'], df['price'])

plt.title('Зависимость стоимости дома от состояния')
plt.xlabel('Состояние дома')
plt.ylabel('Стоимость дома');
```

Чем лучше состояние дома, тем выше его стоимость, но, как видно из диаграммы, дома в идеальном состоянии не оказывали существенного влияния на цену.

```python
plt.figure(figsize=(10, 5))

sns.boxplot(x=df['price'], y=df['waterfront'].astype('str'), whis=1.5)

plt.title('Зависимость стоимости дома от наличия вида на набережную')
plt.xlabel('Стоимость дома')
plt.ylabel('Вид на набережную');
```

В случае наличия вида на набережную стоимость дома увеличиватеся. <br>Так диапазон цен домов в отсутствием вида на набережную доходит до 1,3 млн. с медианной ценой = 0,4 млн, а стоимость домов с наличием такого вида доходит до 3,9 млн с медианной ценой = 1,4 млн.

```python
sns.jointplot(x=df['sqft_living'], y=df['price'], kind='reg');

plt.xlabel('Жилая площадь')
plt.ylabel('Стоимость дома');
```

Цена на дом напрямую зависит от количества жилой площади.

```python
plt.bar(df['bathrooms'], df['price'])

plt.title('Зависимость стоимости дома от количества ванных комнат')
plt.xlabel('Количество ванных комнат')
plt.ylabel('Стоимость дома');
```

Чем больше в доме ванных комнат, тем он дороже стоит.

```python
plt.figure(figsize=(10, 5))

sns.boxplot(x=df['price'], y=df['view'].astype('str'), whis=1.5)

plt.title('Зависимость стоимости дома от его вида')
plt.xlabel('Стоимость дома')
plt.ylabel('Вид дома');
```

Чем лучше выглядит дом, тем дороже он стоит.
