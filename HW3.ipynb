### Задание 1.2
Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновации дома и годом постройки дома. <br>Если реновации дома не было, то в новом признаке поставьте 0.

````python
df['yr_renovated'] = df['yr_renovated'] - df['yr_built']
df['yr_renovated'] = df['yr_renovated'].apply(lambda x: x if x > 0 else 0)
````

## Задача 3
### Задание 3.1
Найдите среднюю стоимость домов в зависимости от количества спален.
<br>Отсортируйте от меньшей стоимости к большей.

````python
df.groupby('bedrooms').agg({'price': 'mean'}).sort_values('price')
````

### Задание 3.2
Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома.


````python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
````

### Задание 3.3
Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида.

````python
df.pivot_table(index='waterfront',
               columns='view',
               values='id',
               aggfunc='count',
               fill_value='0'
               )
````

### Задание 3.4
